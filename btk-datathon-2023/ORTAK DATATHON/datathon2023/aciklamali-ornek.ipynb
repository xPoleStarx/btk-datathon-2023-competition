# Örnek Bir Sınıflandırma Kodu

İlk önce dosyalarımızı okuyalım. Bunun için pandas kütüphanesinin read_csv fonksiyonundan yararlanacağız. Bu fonksiyon ile csv formatı verileri DataFrame objeleri haline getireceğiz.

```python
import pandas
egitim_verisi = pandas.read_csv("/kaggle/input/datathon2023/train.csv")
test_verisi = pandas.read_csv("/kaggle/input/datathon2023/test_x.csv")
```

Şu anda train.csv formatındaki dosyayı okuduk ve egitim_verisi adındaki bir DataFrame'e döndürdük. Aynı şekilde test_x.csv dosyasını da test_verisi şeklinde bir DataFrame'e döndürdük. Bunlara bir göz atalım.

```python
egitim_verisi
```

```python
test_verisi
```

Yukarıda görüleceği gibi egitim_verisi verisinde 14 öznitelik bulunmasına karşı test_verisi tarafında 13 öznitelik bulunuyor. egitim_verisi kısmında kullanılan bir öznitelik (Öbek İsmi) hedef özniteliğimizdir. Diğer öznitelikleri kullanarak "Öbek İsmi" adındaki özniteliği en iyi şekilde anlamlandıran ve tahmin eden modeli yaratmamız gerekiyor. Modeli egitim_verisi üzerinde eğitip, test_verisi üzerinde tahminleme yapacağız.

Modelleme yaparken değerlerin nümerik olması gerektiğinden Kategorik veriler üzerinde işlem yapılması gerekir. Örnek olması açısından burada Erkek değerlere 1, Kadın değerlere 0 diyelim.

```python
egitim_verisi["Cinsiyet"] = egitim_verisi["Cinsiyet"].replace(["Erkek"], 1)
egitim_verisi["Cinsiyet"] = egitim_verisi["Cinsiyet"].replace(["Kadın"], 0)
```

```python
egitim_verisi
```

Başka bir işlem ise One hot encoding yöntemi, diğer özniteliklere de onu uygulayalım.

```python
egitim_verisi = pandas.get_dummies(egitim_verisi, columns=[
                                   'Yaş Grubu', 'Medeni Durum', 'Eğitim Düzeyi', 'Yaşadığı Şehir', 'İstihdam Durumu', 'Eğitime Devam Etme Durumu', 'En Çok İlgilendiği Ürün Grubu'])
```

```python
egitim_verisi
```

Bu işlem sayesinde veri modellenebilir hale dönüşmüş oldu. Şimdi test_verisini de aynı aşamalardan geçirmemiz gerekir.

```python
test_verisi["Cinsiyet"] = test_verisi["Cinsiyet"].replace(["Erkek"], 1)
test_verisi["Cinsiyet"] = test_verisi["Cinsiyet"].replace(["Kadın"], 0)
test_verisi = pandas.get_dummies(test_verisi, columns=['Yaş Grubu', 'Medeni Durum', 'Eğitim Düzeyi',
                                 'Yaşadığı Şehir', 'İstihdam Durumu', 'Eğitime Devam Etme Durumu', 'En Çok İlgilendiği Ürün Grubu'])
```

Şimdi sınıflandırma modeli oluşturup test_verisi indekslerinin "Öbek İsmi" kolonunu yaratmaya çalışalım.

```python
from sklearn.neighbors import KNeighborsClassifier
egitim_verisi_hedefsiz = egitim_verisi.drop(['Öbek İsmi'], axis=1)
model = KNeighborsClassifier()
model.fit(egitim_verisi_hedefsiz, egitim_verisi["Öbek İsmi"])
tahmin_sonuclari = model.predict(test_verisi)
```

Burada rastgele bir sınıflandırıcı seçerek test_verisi değerlerinin Öbek İsimlerini tahmin ettik ve tahmin_sonuçları içerisinde tuttuk.

```python
tahmin_sonuclari
```

```python
veri = {
    'id': range(2340),
    'Öbek İsmi': tahmin_sonuclari
}
Tahmin_Dataframe = pandas.DataFrame(veri)
Tahmin_Dataframe.set_index('id', inplace=True)
Tahmin_Dataframe
```

Yapılan bu tahminleri csv formatında indirip submit etmemiz için.

```python
Tahmin_Dataframe.to_csv('submission.csv')
```
