```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

# Önce modeli eğittiğiniz bölümü tekrarlayarak modeli oluşturun ve eğitin
not_aiming = train_df.drop(['Öbek İsmi'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(
    not_aiming, train_df["Öbek İsmi"], test_size=0.33, random_state=2)

model = XGBClassifier()
model.fit(X_train, y_train)

# Eğitilen modeli test verileri üzerinde değerlendirin
test_predictions = model.predict(X_test)

# Elde edilen tahmin sonuçlarını gerçek etiketlerle karşılaştırarak accuracy hesaplayın
accuracy = accuracy_score(y_test, test_predictions)
print("Accuracy:", accuracy)

# Şimdi test_df veri çerçevesini kullanarak tahmin yapabilirsiniz
# test_predictions = model.predict(test_df)

# Elde edilen tahmin sonuçlarını bir DataFrame'e dönüştürerek inceleme yapabilirsiniz
submission_data = {
    "id": range(0, len(test_predictions)),
    "Öbek İsmi": [f"obek_{cluster+1}" for cluster in test_predictions]
}

submission_df = pd.DataFrame(submission_data)

# Oluşturulan DataFrame'i CSV dosyasına kaydedebilirsiniz
submission_df.to_csv("submissionXGBoost.csv", index=False)
```
